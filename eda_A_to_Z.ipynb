# EDA (Exploratory Data Analysis)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  import data set
df = sns.load_dataset('titanic')
df.head()
```

```python
# summarize
df.describe() # summary statistics of numerical variables
```

```python
df.columns
```

```python
df.drop(columns="deck", inplace=True)
```

```python
df.shape
```

```python
#visualize  and explore 
sns.countplot(data=df, x="survived") # cannot give y-axis because it is already captured
```

```python
sns.barplot(data=df, x="survived", y="age", hue="who") # x-axis categorical with less unique value and y-axis numerical 
```

```python
sns.countplot(x="pclass", data=df) # on categorical variable
```

```python
# data distribution
sns.histplot(data=df, x="age", bins=5) # more bins gives more clarification
```

```python
df['age'].hist(bins=30, color='darkred', alpha=1) # alpha used for opacity
```

```python
sns.boxplot(data=df, y="age")
```

```python
sns.boxplot(data=df, y='age', x='survived' , hue='sex') # one variable should be categorical and other on should be numerical
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='pclass') # both have numerical values
# plt.ylim(0, 300)
# y_limit = (0, 300)
# plt.ylim(*y_limit)
# bringing data at the same scale is called normalization
```

```python
sns.pairplot(df, hue='survived')
```

```python
# stacked bar char/plot for sex
df.groupby(['pclass','sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# stacked bar char/plot for who
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y='fare')
```

```python
# swarm plot
sns.swarmplot(data=df, x='survived', y='age',hue='who')
```

```python
# point plot
sns.catplot(data=df, x='pclass', y='fare', hue='embarked', kind='point')
```

```python
# relation of two numeric variables
sns.lmplot(data=df, x='age', y='fare', hue='pclass')
```

```python
# relation of two numeric variables
filtered_data = df.query('(fare <= 301)')
sns.lmplot(data=filtered_data, x='age', y='fare')
```

```python
sns.lmplot(data=df[df['fare']<300], x='age', y='fare')
```

```python
# how to add column and make a age group (data grouoing)
df['age_groups'] = pd.cut(df['age'], [0,10,20,30,50,80], 
                labels=['kids','child','youth','middle age','old'])
df.head()
```

```python
df.columns
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```
